# Splitting the Data into Training and Testing

```python
onion_big=dataframe.iloc[:,0]
```

```python
train=onion_big.iloc[:int(0.7*onion_big.shape[0])]
test=onion_big.iloc[int(0.7*onion_big.shape[0]):]
```

```python
train_prices=train.values.reshape(-1,1)
test_prices=test.values.reshape(-1,1)
```

```python
# Scaling our data to perform computations in a fast and accurate manner.

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(train_prices)
scaled_train = scaler.transform(train_prices)
scaled_test = scaler.transform(test_prices)
```

```python
scaled_train
```

```python
# Processing to a time series generation.
from keras.preprocessing.sequence import TimeseriesGenerator

n_input = 7
n_features = 1
generator = TimeseriesGenerator(scaled_train,
                                scaled_train,
                                length=n_input,
                                batch_size=1)
```

```python
X,y=generator[0]
print(f"Given the array: \n{X.flatten()}")
print(f"Predict this y:\n{y}")
```

```python
X.shape #(row,column,no of features)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# define model

model=Sequential()
model.add(LSTM(100,activation='relu',input_shape=(n_input,n_features)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
#fit model
model.fit(generator,epochs=25)
```

```python
loss_per_epoch=model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch=scaled_train[-7:] # scaled_train[-n_input]
```

```python
last_train_batch=last_train_batch.reshape((1,n_input,n_features))

# Prepare the last training batch for prediction
# last_train_batch = scaled_train[-n_input:].reshape((1, n_input, n_features))
```

```python
test_predictions=[]

first_eval_batch=scaled_train[-n_input:]
current_batch=first_eval_batch.reshape((1,n_input,n_features))

for i in range(len(test)):
    # get the prediction value for the first batch
    
    current_pred=model.predict(current_batch)[0]
    
    # append the prediction into the array
    test_predictions.append(current_pred)
    
    #use the prediction to update the batch and remove the first value
    current_batch=np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```
